```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = "unicode_escape")
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

Data cleaning


```python
#removing null columns(status,unnamed1)
df.drop(["Status","unnamed1"],axis=1, inplace = True)
```

```python
#checking null values
pd.isnull(df).sum()
```

```python
# deleting null values from column(amount)
df.dropna(inplace=True)
```

```python
#changing databtype of amount column
df["Amount"]=df["Amount"].astype("int")
```

```python
df["Amount"].dtype
```

```python
df.columns
```

```python
# rename (Marital_Status) to (Shade)
#df.rename(columns={"Marital_Status":"Shadi"})
#df.head()
```

```python
# rename (shadi) to (Marital_status)
df.rename(columns={"shadi":"Marital_Status"}, inplace = True)
df.head()
```

```python
#describe (will give mean median etc)
df.describe()
```

```python
# describe specific columns
df[['Age','Orders']].describe()
```

EDA (Explorattory Data Analysis)


```python
df.columns
```

```python
ax = sns.countplot(x="Gender",data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs Gender
sales_data = df.groupby(["Gender"],as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False)
sns.barplot(x="Gender", y="Amount", data= sales_data)
```

Note -> From the above we can see that most of the buyers are females and even the purchasing of the female is greater than men

```python
df.columns
```

Age Group

```python
ax = sns.countplot(x= "Age Group", data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs Age_group
sales_data = df.groupby(["Age Group"] , as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False)
sns.barplot(x="Age Group", y ="Amount", data = sales_data)
```

Note -> From the above graph we can see that most of the buyers are from age group btw 26-35.

```python
ax = sns.countplot(x="Shadi", data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount vs married or un_married people
sales_data =df.groupby(["Shadi","Gender"], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={"figure.figsize":(10,5)})
sns.barplot(x="Shadi",y="Amount", data = sales_data, hue ="Gender")
```

Note -> From the above graph we can see that most of the buyers are married(women) and ther have high purchasing power.

```python
ax = sns.countplot(x="State",data = df)
sns.set(rc={"figure.figsize":(35,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#state vs order
sales_state = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by=["Orders"],ascending= False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
ax=sns.barplot(x="State",y ="Orders",data = sales_state)
```

```python
# state vs amount
sales_state = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by=["Amount"],ascending= False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
ax=sns.barplot(x="State",y ="Amount",data = sales_state)
```

note-> From the above data we can see that the we have highest order and purchase from Uttar Pradesh.

```python
ax = sns.countplot(x="Occupation", data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#occupation vs amount
sales_data = df.groupby(["Occupation", "Gender"], as_index=False)["Amount"].sum().sort_values(by=["Amount"], ascending=False)
sns.barplot(x="Occupation", y="Amount", data=sales_data, hue="Gender")
```

```python
ax=sns.countplot(x="Product_Category", data = df)
sns.set(rc={"figure.figsize":(33,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Product_Category vs order
sales_data = df.groupby(["Product_Category", "Gender"], as_index=False)["Orders"].sum().sort_values(by=["Orders"], ascending=False)
sns.barplot(x="Product_Category", y="Orders", data=sales_data, hue="Gender")
```

```python
#product category vs amount
sales_data = df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by=["Amount"], ascending=False)
sns.barplot(x="Product_Category", y="Amount", data=sales_data)
```
